### Assessing Data

```python
for df_info, df in data.items():
  print('\n', data[df_info].info())
```

```python
duplicates = {}
for df_info, df in data.items():
    duplicates[df_info] = df.duplicated().sum()
    print(df_info, '\n', duplicates[df_info], '\n')
```

```python
null = {}
for df_info, df in data.items():
    null[df_info] = df.isna().sum()
    print(df_info, '\n', null[df_info], '\n')
```

```python
for df_info, df in data.items():
    # Menghitung persentase nilai yang hilang untuk setiap kolom
    percentage_missing = (df.isna().sum() / len(df)) * 100
    null[df_info] = percentage_missing
    print(df_info, '\n', null[df_info], '\n')
```

### Cleaning Data

```python
for df_info, df in data.items():
    data[df_info] = df.dropna()
    print(df_info, '\n', data[df_info], '\n')
```
